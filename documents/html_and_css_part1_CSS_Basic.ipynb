###<font color="blue">***Providing CSS:***</font>
---

Three different ways to provide style:

1. Embedded stylesheet: Where we embed the style in the HTML document.
2. External stylesheet: Which means write the style in a separate files.
3. Inline style: We can apply style directly to a perticular element.

**Example of Embedded stylesheet:**
Let's sy we have a 2 paragraph elements.
Now we set style rule for p and set the the color yelow.
So in forntend, the both paragraphs are blue.
Now what we have here is embedded stylesheet.
```
  <!DOCTYPE html>
  <html lang="en">
    <head>
      <meta charset="UTF-8" />
      <meta name="viewport" content="width=device-width, initial-scale=1.0" />
      <title>Document</title>
      <style>
        p {
          color: blueviolet;
        }
      </style>
    </head>
    <body>
      <p>Lorem ipsum dolor sit amet consectetur.</p>
      <p>Lorem ipsum dolor sit amet consectetur.</p>
    </body>
  </html>
```
Problems with embedded stylesheet:
1. This is not scalable: Let's say if we want to have same color on the other page then we need to copy the same style for each pages. And the if we need to change the color then we need to change the style every pages.
2. The second problem is this approch violate the important principle of computer science called separation of concerns. So there two concerns one html and another one is css so these are needed to be separated.

**Example of External Stylesheet:**
This is where we use the external stylesheet. So we need to move the style rules except the style tag to a separate file which extension must be css.

So let's created a file in root directory called style.css and move the style rule there and save the change.

style.css
```
  p {
    color: blueviolet;
  }
```

Then add link element on the html file to link the external stylesheet.
The link element has two attributes: rel, short from of relationship which determines what type of resource we are linking. In this case it is stylesheet. The second attribute is, href which is the path of the styesheet file that could be absolute/relative path.

index.html
```
  <link rel="stylesheet" href="/style.css" />
```
Now save the change and open this in server and see the same result as we got when the style was embedded.

Overwrite the external stylesheet rule:
Sometime we need to overwrite the external style by using the embedded style. So we just add style and set rule for the paragraph.
```
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
    <link rel="stylesheet" href="/style.css" />
    <style>
      p {
        color: black;
      }
    </style>
  </head>
```

Inline style:
We use style directly to the html element. Let's out of these two paragraphs we want to make the first pragraph blue. So we add style attribute and set the color blue inside the p element.
```
  <body>
    <p style="color: blue">Lorem ipsum dolor sit amet consectetur.</p>
    <p>Lorem ipsum dolor sit amet consectetur.</p>
  </body>
```
We can also set multiple rules like;
```
  <body>
    <p style="color: blue; font-weight: bold">
      Lorem ipsum dolor sit amet consectetur.
    </p>
    <p>Lorem ipsum dolor sit amet consectetur.</p>
  </body>
```
Note: this inline style violates the separation of concern principle. but how can we make the first p blue and bold really? There is a better way that does not violate the separation of concern principle. Give the first p an unique identifier id name like id="first" then in style select the id with #then_the_name_of_the_id then set the rule.
```
  <html lang="en">
    <head>
      <meta charset="UTF-8" />
      <meta name="viewport" content="width=device-width, initial-scale=1.0" />
      <title>Document</title>
      <link rel="stylesheet" href="/style.css" />
      <style>
        p {
          color: black;
        }

        #first {
          color: blue;
          font-weight: bold;
        }
      </style>
    </head>
    <body>
      <p id="first">Lorem ipsum dolor sit amet consectetur.</p>
      <p>Lorem ipsum dolor sit amet consectetur.</p>
    </body>
  </html>
```

To do this, go to google.com and search for normalize.css and download the latest version of the stylesheet from here: https://necolas.github.io/normalize.css/

  <body>
    <h2 class="highlight">Heading for the h2.</h2>
  </body>
```
Stylesheet:
```
  h2 {
    color: blue;
  }

  .highlight {
    color: blueviolet;
  }
```
Now both these rules target to the same h2 element. So what rule the browser apply? This is where Selector Specificity(weight) comes to play.

Each selector has specificity or weight. If multiple rules target to the same element the browser will apply the rule that is more specific.

**Specificity triangle:**
![picture](https://drive.google.com/uc?id=1RuDlpb8ZI8vPrVBYcSxOu5neQPbu9XnS)

Look at this triangle. At the top of the triangle we have;
1. **ID selector:** Because as we know, we can not have multiple elements with same id. So the id selector is very specific it can only target a single element.
2. **Class and attribute selectors:** These selectors are less specific, because we can have multiple elments with same class or attribute.
3. **Element selector:** These are not specific because we have can have multiple elements of a givev type.

**Examples:**

1. **class > type:** So at the above rules, which one is more espcific you think? The answer is, second rule. Because we may have lots of h2 elements but not every h2 have class highlight. So class selector is more specific.

2. **id > class:** Let's add another selector id to the h2 element and set the third rule for the h2 element.
```
  <body>
    <h2 id="product" class="highlight">Heading for the h2.</h2>
  </body>
```
Stylesheet:
```
  #product {
    color: pink;
  }
```
Now which rule should apply? The third one, the id selector one. Becasue id is more specific.

3. **Selectors with same specificity:** What if selectors have same specificity or weight? Let's duplicate the third rule and change the color to something else let's ay black. So which rule should apply in this case?
```
  #product {
    color: pink;
  }

  #product {
    color: black;
  }
```
The rule that comes last will be applied, so the color of the text is black now.


**VScode feature to see weight:**
On the stylesheet, when you hover on the h2 selector a popup will appear and shows the selector specificity like this way bellow:

  ![picture](https://drive.google.com/uc?id=1lBANbqtYkIhNcUwzc1B0IGJ-_3zzsEdo)

Here, the numbers inside the parenthesis indicates the weight, the first digit represents ID, the second one represents Class, and the last one represents element selector. In this case, since this rule is for the h2 element so it shows (0,0,1) means there is no id nor class but 1 element. And the specificity in this case is, 1.

Now hover on the class selector. You see the specificity is(0,1,0) which is 10. That indicates more specific than eelement selector.

Similarly hover on the id selector and it shows (1,0,0) mean 100 and it is most specific selector.

**Overwrite the specificty rule and apply style:**
At this point, the most specificity one is id selector and, since the rule is set to clolor: pink so currently the h2 is showing pink. But if we want to overwirte the rule with the less specificity one I mean the apply the rule for the h2 element.

There are two soloutions:
>1. apply 'important' keyword: Add '!important' keyword inside the rule
```
  h2 {
    color: blue !important;
  }
```
Now the text shows bule instead of pink. Because by using the important keyword it overwrite the rule. But using this is not a good practice. Because it make the code messy and hard to maintain.

>2. Make the selector more specificity by adding selector: If we want to show the text in other color then we need to write the selector in a way so the weight/specificity become more than the current one, which is (1,0,0). Than mean we need atleast an id, and a class/element selector.
```
  .highlight#product {
    color: black;
  }
```
Now the specificity is (1,1,0) so this rule will apply and the color will show black now.

**Another way to select:**
Let's give the article element an id and select by that id and class this is how the specificity can be more and the rule will be applied to the element.
```
  <article class="featured">
    <h2 id="product" class="highlight">Heading for the h2.</h2>
  </article>
```
Stylesheet:
```
  .featured #product {
    color: black;
  }
```
Here the specificity is 110 so it is more specific.

###<font color="blue">***Inheritance:***</font>
---

One of the important concepts of CSS is inheritance. Which means some CSS propeties by default inherite values from their parents. Let's go through few examples.

**Example 1:**
1. Let's add a paragraph and some random text inside.
2. Than wrap a word inside a strong element.
3. Than in stylesheet, select by element p and set the rule color: bule
html file:
```

  <section>
    <article class="box"></article>
  </section>
```

**Named colors:**
Now in stylesheet select the element by class .box and set the rule for styleing the box.
```
  .box {
    width: 200px;
    height: 200px;
    background-color: blue;
  }
```
Here the 'blue' name of the color is the usage of Named color.

**RGB:**
In stylesheet, let's replace the named color with RGB value. Go to google and search for 'color picker' google has a built-in color picker.
![picture](https://drive.google.com/uc?id=1hjfS2LoGUqZOPQGlRg4Xyw839pw88a8L)

In this color picker, when we are miving the picker circle around the color box the values of HEX, RGB, HSL are changing. Let's take look at the values one by one.

RGB is short for Red, Green and Blue and these three values represent the degree of red, green and blue. Each value can be number between 0 to 255. So the colore in the picker represent the combination of red, green and blue. So we can copy the values and  to set the rule, use the rgb() function and pass the vlaues for the background-color property.
```
  .box {
    width: 200px;
    height: 200px;
    background-color: rgb(181, 86, 31);
  }
```

**rgbc() function:**
We have another function called rgba(), which means alpha. With this function we can supply an extra value. And the value can be a decimal number between 0 and 1. 0 represents completly transparent and 1 means completly opaque. So if we want to make the box semi transparent then supply the value 0.5
```
  .box {
    width: 200px;
    height: 200px;
    background-color: rgba(181, 86, 31, 0.5);
  }
```

**HEX:**
We also represent it by using hexadecimal system. Hexadecimal is one of the many numerical systems.
1. Binary(0,1): That contain only 0 and 1
2. Decimal(0-9): That contains the digit between 0 to 9
3. Hexadecimal(0-9, a-f): That contains the digit 0 to 9 and the letter a to f.
We can represents any number of the word using any of these systems.

The HEX value has 6 digits and every 2 digits represent the numbers/value of RGB system. Means the first 2 digits in HEX system represents the RED value in the RGB system, the second 2 digits represent the value of Green, and the last 2 digits represents the Blue in rgb suystem.

Now copy the HEX value and set the rule for background-color. And this results exactly same as when we used RGB value. Biut the limitation of HEX system is it has no transparency channel.
```
  .box {
    width: 200px;
    height: 200px;
    background-color: #b5561f;
  }
```

**HSL:**
Another way to represent color is HSL system. HSL has three values(hue, saturation, lightness) Hue is a degree on the color wheel from 0 to 360. 0 is red, 120 is green, and 240 is blue. Saturation is a percentage value. 0% means a shade of gray, and 100% is the full color. Lightness is also a percentage value. 0% is black, and 100% is white.

We can use this HSL values to make diffenrt shades of colors seletion the color and the adjusting the saturation and lightness. To use this, copy the HSL values pass to the hsl() for the background-color prperty while setting the style rule. But 1 thing, we need to remove the degree sign from the color value.
```
  .box {
    width: 200px;
    height: 200px;
    background-color: hsl(22, 71%, 42%);
  }
```
**hsla() function:**
Same as RGB, it has transparency channel. So use hsla() function and pass value between 0 and 1 to make tranparency.
```
  .box {
    width: 200px;
    height: 200px;
    background-color: hsla(22, 71%, 42%, 0.5);
  }
```
**Chrome color picker:**
Run your html document in live server from vscode and it open in chrome browser. Now right click on the box and inspect element. Than on the right side you can see the style here you can see the color click on the color icon and open the color picker. From here you can select color and at the same time you can see the changes in the browser. Now you can copy the value from here and apply in the stylesheet.

  ![picture](https://drive.google.com/uc?id=1XGbJodSGpUAsl80Ozn0c25aewGgzrDxU)
```

  .box:nth-child(2) {
    width: 300px;
    height: 300px;
    background-image: linear-gradient(dodgerblue, yellow);
    margin: 10px;
  }
```
The image is look like now: The transition is top to bottom blue to yellow.

  ![picture](https://drive.google.com/uc?id=1rBEKnb5fccRQn3iTU8CJ9Y2Fj-RELZYT)

Now we can change the transition direction various way. To do this we need pass direction value before the first color in the function. Like this:

Transition left to right:
```
  background-image: linear-gradient(to right, dodgerblue, yellow);
```

Transition top left to bottom right:
```
  background-image: linear-gradient(to bottom right, dodgerblue, yellow);
```

Transition by dgree value:
```
  background-image: linear-gradient(45deg, dodgerblue, yellow);
```
We can also specify the position of color in the box:
Like we can say yellow should start 30% of the box space like below way.
```
  background-image: linear-gradient(45deg, dodgerblue, yellow 30%);
```

**Radial gradient:**
Use radial-gradient() and pass two colors to set the rule for the 3rd box.
```
  .box:nth-child(3) {
    width: 200px;
    height: 200px;
    background-image: radial-gradient(dodgerblue, yellow);
    margin: 10px;
  }
```
Here we get the image like, a white circle in center and gardually transition to the outside to yellow. And the center is circle shape because the container is squre shape.

And if we make the box rectangle shape then the center will be ellipse. In this case we can make the center circle we supply circle keyword as the firt argument in the fucntion.
```
  .box:nth-child(3) {
    width: 600px;
    height: 200px;
    background-image: radial-gradient(circle, dodgerblue, yellow);
    margin: 10px;
  }
```

We can also change position of the circle by saying 'at top left' for example
```
  .box:nth-child(3) {
    width: 600px;
    height: 200px;
    background-image: radial-gradient(circle at top left, dodgerblue, yellow);
    margin: 10px;
  }
```

**Online tool to generate gradient and css codes:**
1. Go to google
2. Search for 'gradient generator'
3. Go to: https://cssgradient.io/
4. Make a gradient and copy the code
5. Paste the code in stylesheet
```

  .box:nth-child(10) {
    width: 200px;
    height: 200px;
    background: dodgerblue;
    margin: 10px;
    box-shadow: 10px 10px;
  }
```
And the image we get:

  ![picture](https://drive.google.com/uc?id=10B7QnSNoKq-QC7mDvrdQ10OMNGQxrMDw)

**Third vaue:** Third value is to make the shadow soften.
```
  .box:nth-child(10) {
    width: 200px;
    height: 200px;
    background: dodgerblue;
    margin: 10px;
    box-shadow: 10px 10px 10px;
  }
```

**Fourth value:** The fourth value is the color of the shadow, by default it is balck. Let's change it to grey.
```
  .box:nth-child(10) {
    width: 200px;
    height: 200px;
    background: dodgerblue;
    margin: 10px;
    box-shadow: 10px 10px 10px grey;
  }
```

Now a trick to make the nice shadow effect. The current shadow does not cover the top and left. Whe you use a very soft shadow you can use this tirck, not to supply the horizental and vertical distance just pass zero instead. With this tirck, the shadow will out right under the box and this will have a nice effect of shadow.
```
  .box:nth-child(10) {
    width: 200px;
    height: 200px;
    background: dodgerblue;
    margin: 10px;
    box-shadow: 0 0 30px grey;
  }
```

**Shaodw the text:** Default we have the black text if we add a h1/h2/h3 whatever in our html. Then set a rule in CSS to change the color of the text.
```
  h3 {
    color: white;
  }
